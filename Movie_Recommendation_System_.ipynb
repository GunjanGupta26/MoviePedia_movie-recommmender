```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv("5000tmbd.csv")
credits=pd.read_csv("5000tmdbcredit.csv")
```

```python
merge = movies.merge(credits, on='id')
merge.shape
```

```python
data = merge.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
data.head()
```

```python
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj): 
        l.append(i['name'])
    return l
```

```python
data['genres']=data['genres'].apply(convert)
data['production_companies']=data['production_companies'].apply(convert)
data['spoken_languages']=data['spoken_languages'].apply(convert)
data['keywords']=data['keywords'].apply(convert)
data['cast']=data['cast'].apply(convert)
```

```python
data["genre"] = [','.join(map(str, l)) for l in data['genres']]
data["casts"] = [','.join(map(str, l)) for l in data['cast']]
```

```python
data[['actor1', 'actor2', 'actor3','actor4']] = data['casts'].str.split(',',n= 3, expand=True)
data[['genre1', 'genre2', 'genre3','genre4']] = data['genre'].str.split(',',n= 3, expand=True)
data[['year','date']] = data['release_date'].str.split('-',n= 1, expand=True)
```

```python
data['genre2'].fillna(data['genre1'], inplace = True)
data['genre3'].fillna(data['genre1'], inplace = True)
data['genre4'].fillna(data['genre1'], inplace = True)

data['actor2'].fillna(data['actor1'], inplace = True)
data['actor3'].fillna(data['actor1'], inplace = True)
data['actor4'].fillna(data['actor1'], inplace = True)
```

```python
data["o+g"] = data["genre1"].map(str)+ ' ' +  data["genre2"] + ' ' + data["overview"] 
data.head()
```

```python
data = data.drop(columns=['cast','tagline'])
data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df=3,  max_features=None,
            strip_accents='unicode', analyzer='word',
            ngram_range=(1, 3),
            stop_words = 'english')
            
tfv_matrix = tfv.fit_transform(data['overview'].values.astype('U'))
tfv_matrix1 = tfv.fit_transform(data['o+g'].values.astype('U'))
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
model = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
model1 = sigmoid_kernel(tfv_matrix1, tfv_matrix1)
```

```python
data1 = pd.Series(data.index, index=data['original_title']).drop_duplicates()
```

Recommend on the basis of Overview

```python
def recommend_overview(title, s=model):
    idx = data1[title]
    sig_scores = list(enumerate(s[idx]))
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    sig_scores = sig_scores[1:6]
    movie_indices = [i[0] for i in sig_scores]
    return data['original_title'].iloc[movie_indices]
```

```python
print(recommend_overview('Transformers: Age of Extinction'))
```

Recommend on the basis of Overview and Genre

```python
def recommend_overview_genre(title, s=model1):
    idx = data1[title]
    sig_scores = list(enumerate(s[idx]))
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    sig_scores = sig_scores[1:6]
    movie_indices = [i[0] for i in sig_scores]
    return data['original_title'].iloc[movie_indices]
```

```python
print(recommend_overview_genre('Transformers: Age of Extinction'))
```
